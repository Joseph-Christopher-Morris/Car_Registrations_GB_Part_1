```python
selected_makes = ["VOLKSWAGEN", "FORD", "TOYOTA", "VAUXHALL", "MERCEDES"]
```

```python
cars_2024Q2 = registrations[
    (registrations['Make'].isin(selected_makes)) &
    (registrations['2024Q2'] > 0)
]
```

```python
registrations_per_make = cars_2024Q2.groupby('Make')['2024Q2'].sum()
registrations_per_make = registrations_per_make.reindex(selected_makes)
```

```python
registrations_per_make = cars_2024Q2.groupby('Make')['2024Q2'].sum()
registrations_per_make = registrations_per_make.reindex(selected_makes)

# Define chart labels and values
labels = registrations_per_make.index
values = registrations_per_make.values

# Brand-specific colors
colors = {
    "FORD": "#0055A4",
    "VOLKSWAGEN": "#C0C0C0",
    "VAUXHALL": "#32CD32",
    "MERCEDES": "#4D4D4D",
    "TOYOTA": "#EB0A1E",
}

color_list = [colors.get(label.upper(), "grey") for label in labels]

# Function to show percentage and count
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n({val:,})"
    return my_autopct

# Create pie chart
fig, ax = plt.subplots(figsize=(8, 6), facecolor="#101031")

wedges, texts, autotexts = ax.pie(
    values,
    labels=labels,
    colors=color_list,
    autopct=make_autopct(values),
    startangle=90,
    pctdistance=0.8,
    textprops={'fontsize': 10}
)

for text in texts:
    text.set_color("white")
for autotext in autotexts:
    autotext.set_color("black")

ax.set_title("Car Registrations â€“ GB 2024 Q2", fontsize=14, fontweight='bold', color="white")
plt.figtext(0.5, 0.01, "Source: Department for Transport (DfT), UK | Q2 2024", ha="center", fontsize=9, color="white")
ax.axis('equal')
plt.tight_layout()
plt.show()
```

```python
model_keywords = {
    "1 Series": "1",
    "3 Series": "3",
    "5 Series": "5",
}

quarters = [f"{year}Q2" for year in range(2014, 2025)]

# Create an empty DataFrame to hold the results
series_data = pd.DataFrame()
```

```python
for label, keyword in model_keywords.items():
    filtered = registrations[
        (registrations['Make'].str.upper() == "BMW") &
        (registrations['Model'].str.upper().str.contains(keyword))
    ]
    summed = filtered[quarters].sum()
    series_data[label] = summed

# Transpose so models become rows
series_data = series_data.T
```

```python
series_data.index = series_data.index.str.title()

line_colors = {
    "1 Series": "#81C4FF",
    "3 Series": "#71CC51",
    "5 Series": "#E7222E"
}

fig, ax = plt.subplots(figsize=(10, 6), facecolor="#101031")
ax.set_facecolor("#101031")

for model in series_data.index:
    y_values = series_data.loc[model]
    
    # Draw the line
    ax.plot(
        quarters,
        y_values,
        label=model,
        linewidth=3,
        color=line_colors.get(model, "#FFFFFF")
    )

    # Add white label at 2024Q2 point
    value_2024 = y_values["2024Q2"]
    ax.text(
        x=quarters[-1],
        y=value_2024,
        s=f"{value_2024:,}",
        va='bottom',
        ha='left',
        fontsize=10,
        color="white",  # <<< White text
        fontweight='bold'
    )

# Chart styling
ax.set_title("BMW Registrations (2014â€“2024) GB: 1, 3, and 5 Series", fontsize=14, fontweight='bold', color="white")
ax.set_xlabel("Year", fontsize=12, color="white")
plt.ylabel("Registrations measured in 100,000's (0.1's)", color='white')
ax.tick_params(colors='white')
ax.legend(facecolor="#101031", labelcolor='white', fontsize=10)
plt.xticks(rotation=45)
plt.figtext(0.5, 0.0005,
            "Source: Department for Transport (DfT), UK | Q2 figures only",
            ha="center", fontsize=9, color="white")
plt.tight_layout()
plt.show()
```

```python
exclude_vans = ["TRANSIT", "VIVARO", "SPRINTER", "TRAFFIC", "KANGOO", "BERLINGO", "COMBO", "DOBLO"]

cars_only = registrations[
    registrations["GenModel"].notna() &
    registrations["2024Q2"].notna() &
    ~registrations["GenModel"].str.upper().str.contains("|".join(exclude_vans), na=False)
]
```

```python
df_model = cars_only.groupby("GenModel")["2024Q2"].sum().reset_index().sort_values(by="2024Q2", ascending=False).head(10)

sizes = df_model["2024Q2"].tolist()
genmodels = df_model["GenModel"].tolist()

# Color mapping
norm = mcolors.Normalize(vmin=min(sizes), vmax=max(sizes))
cmap = plt.get_cmap("RdYlGn_r")
colors = [cmap(norm(value)) for value in sizes]
```

```python
labels = []
text_colors = []

for i, val in enumerate(sizes):
    label = f"{genmodels[i].upper()}\n{val:,.0f} cars"
    r, g, b, _ = colors[i]
    brightness = 0.299 * r + 0.587 * g + 0.114 * b
    text_color = "black" if brightness > 0.6 else "white"
    labels.append(label)
    text_colors.append(text_color)
```

```python
fig = plt.figure(figsize=(14, 18), facecolor="#0a0a19")
ax = plt.gca()

squarify.plot(
    sizes=sizes,
    label=labels,
    color=colors,
    alpha=0.95,
    ax=ax,
    text_kwargs={'fontsize': 16, 'weight': 'bold'}
)

# Manually apply text colors
for text, color in zip(ax.texts, text_colors):
    text.set_color(color)

plt.axis("off")
plt.title("Most Registered Car Models in GB â€“ Q2 2024", fontsize=26, weight='bold', color='white')

# âœ… Fix: Attach colorbar to the current Axes
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, aspect=30)
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='white')

plt.tight_layout()
plt.show()
```
